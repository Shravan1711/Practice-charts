Donut chart


```
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_7
```

```
name=['group A','group B','group C','group D']
size=[12,11,3,30]
circle=plt.Circle((0,0),0.7,color='white')
```

```
plt.pie(size,labels=name,colors=Pastel1_7.hex_colors)
p=plt.gcf()
p.gca().add_artist(circle)
plt.show()
```

Background Map

```
import pandas as pd
```

```
data=pd.DataFrame({'lon':[-58,2,145,30.32,-4.03,-73.57,36.82,-38.5],
                   'lat':[-34,49,-38,59.93,5.33,45.52,-1.29,-12.97],
                   'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador'],
                   'value':[10, 12, 40, 70, 23, 43, 100, 43]},dtype=str)
data
```

```
import folium
m=folium.Map(location=[20,0],titles='openstreetmap',zoom_start=2)
m
```

```
for i in range(0,len(data)):
  folium.Marker(location=[data.iloc[i]['lat'],data.iloc[i]['lon']],
                popup=data.iloc[i]['name']).add_to(m)
m
```
